```python
# 容错 + 简洁
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bsObj = BeautifulSoup(html.read())
        title = bsObj.h1         # try 语句里面不要写return 语句，写了无法完成预期。
    except AttributeError as e:
        return None
    return title

title = getTitle("http://www.pythonscraping.com/pages/page1.html")
if title is None:
    print("title not exist.")
else:
    print(title)
```

```python
# 通过属性名抓取数据
def getDataByClass(url, tag, attr, attr_name):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bsObj = BeautifulSoup(html.read())
        data = bsObj.find_all(tag, {attr: attr_name})
    except AttributeError as e:
        return None
    return data

data = getDataByClass("https://www.pythonscraping.com/pages/warandpeace.html", "span", "class", "green")
if data is None:
    print("data not exist.")
else:
    for single_data in data:
        print(single_data.get_text())
# get_text() 只保留文字，同时只能使用一次，一次即破坏html所有结构，因而应该在最后使用
```

```python
# 区分子标签和后代标签（文本也算一个子节点）
html = urlopen("http://www.pythonscraping.com/pages/page3.html")
bsObj = BeautifulSoup(html)
for child in bsObj.find("table",{"id":"giftList"}).children:
 print(child)
```

```python
for descendant in bsObj.find("table",{"id":"giftList"}).descendants:
 print(descendant)                      # 有些节点被重复打印
```

```python
# 处理表格数据 ：选择标签行之后的所有行
html = urlopen("http://www.pythonscraping.com/pages/page3.html")
bsObj = BeautifulSoup(html)
for sibling in bsObj.find("table",
{"id":"giftList"}).tr.next_siblings:
 print(sibling)
```

```python
# 获取属性的值
import re
html = urlopen("http://www.pythonscraping.com/pages/page3.html")
bsObj = BeautifulSoup(html)
images = bsObj.findAll("img",
{"src":re.compile("\.\.\/img\/gifts\/img.*\.jpg")})
for image in images:
 print(image["src"])  # 获取标签的src属性
```

```python
html = urlopen("http://www.pythonscraping.com/pages/page3.html")
soup = BeautifulSoup(html)
l = soup.findAll(lambda tag: len(tag.attrs) == 2)
for s_l in l:
    print(s_l, end="\n\n")
# BeautifulSoup 允许我们把特定函数类型当作 findAll 函数的参数。唯
# ⼀的限制条件是这些函数必须把⼀个标签作为参数且返回结果是布尔
# 类型。
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen("http://en.wikipedia.org/wiki/Kevin_Bacon")
bsObj = BeautifulSoup(html)
for link in bsObj.findAll("a"):
 if 'href' in link.attrs:
    print(link.attrs['href'])
```

```python
# 维基百科的六度分析
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
html = urlopen("http://en.wikipedia.org/wiki/Kevin_Bacon")
bsObj = BeautifulSoup(html)
for link in bsObj.find("div", {"id":"bodyContent"}).findAll("a",
 href=re.compile("^(/wiki/)((?!:).)*$")):
 if 'href' in link.attrs:
    print(link.attrs['href'])
```
